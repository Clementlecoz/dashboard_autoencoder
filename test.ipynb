```python
import yfinance as yf
import pandas as pd

# List of companies and their tickers
companies = {
    "Credit_Agricole": "ACA.PA",
    "JPMorgan_Chase": "JPM",
    "TotalEnergies": "TTE.PA",
    "Apple": "AAPL"
}

for name, ticker_symbol in companies.items():
    ticker = yf.Ticker(ticker_symbol)
    quarterly_financials = ticker.quarterly_financials

    # Safely extract desired rows if they exist
    data = {}
    if 'Total Revenue' in quarterly_financials.index:
        data['Total Revenue'] = quarterly_financials.loc['Total Revenue']
    if 'Net Income' in quarterly_financials.index:
        data['Net Income'] = quarterly_financials.loc['Net Income']

    df = pd.DataFrame(data)

    # Save to Excel
    filename = f"{name}_quarterly_financials.xlsx"
    df.to_excel(filename)
    print(f"Saved: {filename}")
```

```python
import yfinance as yf
import pandas as pd

# Load data for Crédit Agricole (ticker on Paris Exchange)
ticker = yf.Ticker("AAPL")

# Retrieve quarterly financial statements
income = ticker.quarterly_financials.T
balance = ticker.quarterly_balance_sheet.T
cashflow = ticker.quarterly_cashflow.T

# Initialize the main DataFrame
df = pd.DataFrame(index=income.index)

# Extract core indicators
df['Revenue'] = income.get('Total Revenue')
df['Net Income'] = income.get('Net Income')

df['Total Assets'] = balance.get('Total Assets')
df['Total Liabilities'] = balance.get('Total Liabilities')
df['Shareholders Equity'] = balance.get('Total Stockholder Equity')

df['Current Assets'] = balance.get('Total Current Assets')
df['Current Liabilities'] = balance.get('Total Current Liabilities')

df['Operating Cash Flow'] = cashflow.get('Total Cash From Operating Activities')
df['CapEx'] = cashflow.get('Capital Expenditures')  # Usually negative

# Derived indicators
df['ROE'] = df['Net Income'] / df['Shareholders Equity']
df['Net Margin'] = df['Net Income'] / df['Revenue']
df['Debt-to-Equity'] = df['Total Liabilities'] / df['Shareholders Equity']
df['Current Ratio'] = df['Current Assets'] / df['Current Liabilities']
df['Free Cash Flow'] = df['Operating Cash Flow'] + df['CapEx']  # CapEx is negative
df['Asset Turnover'] = df['Revenue'] / df['Total Assets']
df['Revenue Growth'] = df['Revenue'].pct_change()

# Optional: Clean up and sort
df.sort_index(inplace=True)
df.dropna(how='all', inplace=True)

# Export to CSV
df.to_csv("credit_agricole_financial_indicators.csv")

print("✅ Data exported successfully to 'credit_agricole_financial_indicators.csv'")
```
