```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset with proper separator and decimal handling
df = pd.read_csv("dataset_unified.csv", sep=';', decimal=',')

# Extract only numeric columns
df_numeric = df.select_dtypes(include=['number'])

# Quick check: Show which columns are numeric and first few rows
print("Numeric columns:", df_numeric.columns.tolist())
print(df_numeric.head())

# Compute correlation matrix
corr_matrix = df_numeric.corr()

# Print the correlation values
print("\nCorrelation Matrix:\n", corr_matrix)

# Plot the heatmap
plt.figure(figsize=(14, 12))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix: Financial & Macroeconomic Variables", fontsize=16)
plt.tight_layout()
plt.show()
```
